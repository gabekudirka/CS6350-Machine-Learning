```python
#Problem 2b plots
bagged_test_errors = np.genfromtxt('test_errs_bt.txt', delimiter=',')
bagged_train_errors = np.genfromtxt('training_errs_bt.txt', delimiter=',')
```

```python
T_ = [1, 2, 2, 5, 5, 5, 10, 20, 25, 25, 50, 50, 100, 100, 100]
T = []
t = 0
for t_ in T_:
    t += t_
    T.append(t)
T
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(10)
plt.plot(T, bagged_train_errors, label='Training Error')
plt.plot(T, bagged_test_errors, label='Testing Error')
plt.title('Bagged Trees Test/Training Error by Iteration')
plt.xlabel('Number of Trees')
plt.ylabel('Error')
plt.legend()
```
